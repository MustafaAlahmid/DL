## Creating Tensors

```python
string = tf.Variable("this is string", tf.string)
number = tf.Variable(100, tf.int16)
Float = tf.Variable(1.33, tf.float64)
```

```python
print(string,
      number,
      Float)
```

## Rank of Tensor

### Rank means the number of dimension of the tensor

```python
rank1_test = tf.Variable("test", tf.string)
rank2_test = tf.Variable([['test'],['test']],tf.string)
```

```python
print(tf.rank(rank1_test))
print(tf.rank(rank2_test))
```
